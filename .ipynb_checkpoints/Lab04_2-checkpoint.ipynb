```python
class Graph:
    """My class for Lab04"""
    def __init__(self, gal_file):

        # get file name
        self.gal_file = gal_file


        # open and read gal file
#         gal_file=r'C:\Users\yangw\PTUA2022\Lab04\Lab\Lab04-1.gal'
        # open and read gal file
        fp = open(gal_file)
        connectivities = fp.readlines()
        fp.close()
        # print(connectivities)
        
        
        self.connectivities = connectivities

        o_d={}
        o_cards={}
        for i,con in enumerate(self.connectivities[1:]):
            print(i)
            print(con)
            if i%2==0:
#                 print(i)
                o_node=int(con.split(' ')[0])
                o_cardi=int(con.split(' ')[1])
                o_cards[o_node]=o_cardi

#                 print('node %s has %s cardi '%(o_node,o_cardi))
            else:
#                 print(i)
                o_connectedTo=[int(c) for c in con.split(' ')]
#                 print('node %s connected to %s'%(o_node,o_connectedTo))
                o_d[o_node]=o_connectedTo
        print(o_d)
        print(o_cards)
        self.o_d=o_d
        self.o_cards=o_cards
    
    
        
    def AvergeConnectivity(self):
        return(np.mean(list(self.o_cards.values())))

    def MaxConnectivity(self):
        maxConns=max(self.o_cards.values())
        maxNodes=[]
        for k,v in self.o_cards.items():
            if v==maxConns:
                maxNodes.append(k)
        return((maxConns,maxNodes))
        
    def Islands(self):
        islands=[]
        for i in range(1,max(self.o_cards.keys())+1):
            if i not in self.o_cards.keys():
                islands.append(i)
        return(islands)
        
      
    
        

if __name__ == '__main__':

    my_gal = Graph(r"C:\Users\yangw\PTUA2022\Lab04\Lab\Lab04-1.gal")
    print('there are islands:%s'%len(my_gal.Islands()))
    print(my_gal.MaxConnectivity())
    print(my_gal.AvergeConnectivity())

    
    
```
