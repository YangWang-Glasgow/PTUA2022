```python
import random
import math
import matplotlib.pyplot as plt

pi = math.pi # our benchmark that we want to estimate using simulation
pi
```

```python
import numpy as np
from functools import reduce
convergencies=[0.01]#, 0.001, 0.0001, 0.00001]
runLim=int(10)
for conv in convergencies:
    draws=[]
    for timeToRun in range(0,1):
        n = 0 # number of points falling in the unit circle
        d = 0 # number of points falling in the unit square
        xs=[]
        ys=[]
        ratios=[]
    
        simulating = True # use as a sentinel
        cnt=0
        while simulating:
            x = random.random()
            y = random.random()
            xs.append(x)
            ys.append(y)
            if x**2 + y**2 <= 1.0:
                n += 1
            d += 1
            ratio = 4 * n * 1./d
            print(ratio)
            ratios.append(ratio)
            if abs(ratio-pi) / pi <= 0.00001:
                print("Draws needed: ", d)
                ds.append(d)
                break
        
        print('for convergens %s'%conv)
        print ('restuls are:%s'%ratio)
        print(' draws:%s'%ds)
        draws.append(ds)
    print('total run:%s'%len(draws))
    print('Drows in 10 run %s'%(draws))
    unpackedDraws=reduce(lambda c, x: c + x, draws, [])

    print('Mean Draws:%s'%np.mean(unpackedDraws))
    print('sdt Draws:%s'%np.std(unpackedDraws))

        

    
```
